```python
import os
import pandas as pd
from datetime import datetime

# Define the input folder and output folder
input_folder = "4_Final_Data"
output_folder = "5_Split_Data"

# Date range (you can customize or remove if not needed)
start_date = datetime(2022, 1, 10)
end_date = datetime(2025, 1, 10)

# Ensure output directories exist
os.makedirs(output_folder, exist_ok=True)
os.makedirs(os.path.join(output_folder, "Train"), exist_ok=True)
os.makedirs(os.path.join(output_folder, "Test"), exist_ok=True)

# Dictionary to track sector-wise company counts
sector_counts = {}
skipped_companies = []

# First pass: Identify companies with valid 3-year data and count sectors
valid_companies = []
for file_name in os.listdir(input_folder):
    file_path = os.path.join(input_folder, file_name)
    if os.path.isfile(file_path) and file_name.endswith(".csv"):
        df = pd.read_csv(file_path, parse_dates=['Date'])
        df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

        # Check for exactly 3 years of trading data (you may adjust as needed)
        if len(df) == 741:
            sector = df['Sector'].iloc[0] if 'Sector' in df.columns and not df['Sector'].isna().all() else None

            if sector:
                valid_companies.append((file_name, sector))
                sector_counts[sector] = sector_counts.get(sector, 0) + 1
            else:
                skipped_companies.append((file_name, "Missing Sector Column"))
        else:
            skipped_companies.append((file_name, "Insufficient Data"))

# Filter sectors with at least 2 companies
valid_sectors = {sector for sector, count in sector_counts.items() if count >= 2}

# Split function for 80-20 split
def split_data_80_20(df):
    df = df.sort_values(by='Date').reset_index(drop=True)
    split_index = int(0.8 * len(df))
    train = df.iloc[:split_index]
    test = df.iloc[split_index:]
    return train, test

# Second pass: Split and save for valid companies in valid sectors
for file_name, sector in valid_companies:
    if sector in valid_sectors:
        file_path = os.path.join(input_folder, file_name)
        df = pd.read_csv(file_path, parse_dates=['Date'])
        df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
        train, test = split_data_80_20(df)
        
        # Save the files
        base_name = os.path.splitext(file_name)[0]
        train.to_csv(os.path.join(output_folder, "Train", f"{base_name}.csv"), index=False)
        test.to_csv(os.path.join(output_folder, "Test", f"{base_name}.csv"), index=False)
    else:
        skipped_companies.append((file_name, "Sector has fewer than 2 companies"))

print(f"Data has been split into 80% Train and 20% Test in '{output_folder}' folder.")
```

```python
import pandas as pd
import numpy as np
import os

def sliding_window_transform(data, window_size=5):
    """
    Converts daily stock data into weekly data using a sliding window approach.
    """
    transformed_data = []
    for i in range(len(data) - window_size - 1):  
        window = data['Close'].iloc[i:i + window_size].values  
        future_price = data['Close'].iloc[i + window_size]  
        return_ratio = (future_price - window[-1]) / window[-1]  
        movement = 1 if future_price > window[-1] else 0  
        transformed_data.append(list(window) + [return_ratio, movement])
    column_names = [f'Day_{i+1}' for i in range(window_size)] + ['Return_Ratio', 'Movement']
    return pd.DataFrame(transformed_data, columns=column_names)

# Paths
input_base_folder = "5_Split_Data"
output_base_folder = "6_Sliding_Window"
os.makedirs(output_base_folder, exist_ok=True)

# Track processed stocks
NIFTY500_stock = {}

# Loop through Train, Test, Validation
for folder in ["Train", "Test"]:
    input_folder = os.path.join(input_base_folder, folder)
    output_folder = os.path.join(output_base_folder, folder)
    os.makedirs(output_folder, exist_ok=True)

    for file in os.listdir(input_folder):
        if file.endswith(".csv"):
            # Rename: Remove '.NS_data' from filename
            company_name = file.replace(".NS.csv", "")
            file_path = os.path.join(input_folder, file)

            df = pd.read_csv(file_path, parse_dates=['Date'])
            df = df.sort_values(by='Date').reset_index(drop=True)

            transformed_df = sliding_window_transform(df, window_size=5)

            output_file = os.path.join(output_folder, f"{company_name}.csv")
            transformed_df.to_csv(output_file, index=False)

            NIFTY500_stock[company_name] = transformed_df
            print(f"âœ… Processed {company_name} in {folder}")

print("\nAll valid companies transformed and saved in standard format.")
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=5000):
        super(PositionalEncoding, self).__init__()
        pe = torch.zeros(max_len, d_model)  # (max_len, d_model)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)  # (max_len, 1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        
        pe[:, 0::2] = torch.sin(position * div_term)  # even indices
        pe[:, 1::2] = torch.cos(position * div_term)  # odd indices
        pe = pe.unsqueeze(0)  # (1, max_len, d_model)
        self.register_buffer('pe', pe)

    def forward(self, x):
        # x shape: (batch_size, seq_len, d_model)
        x = x + self.pe[:, :x.size(1)]
        return x

class LSTT(nn.Module):
    def __init__(self, input_dim=5, model_dim=64, num_layers=2, output_dim=64, nhead=4):
        super(LSTT, self).__init__()
        self.input_proj = nn.Linear(input_dim, model_dim)  # Project input to model_dim
        self.pos_encoder = PositionalEncoding(model_dim)
        encoder_layers = nn.TransformerEncoderLayer(d_model=model_dim, nhead=nhead, batch_first=True)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layers, num_layers=num_layers)
        self.output_proj = nn.Linear(model_dim, output_dim)

    def forward(self, x):
        # x shape: (batch_size, seq_len, input_dim)
        x = self.input_proj(x)                    # (batch, seq_len, model_dim)
        x = self.pos_encoder(x)                   # Add positional encoding
        x = self.transformer_encoder(x)           # Transformer layers
        out = x[:, -1, :]                         # Use last timestep's embedding
        return self.output_proj(out)              # Final projection to output_dim


# Model settings
input_dim = 5
hidden_dim = 64
num_layers = 2
output_dim = 64

# Initialize model
model = LSTT(input_dim=input_dim, model_dim=64, num_layers=num_layers, output_dim=output_dim)


# ðŸ§ª You can optionally load pretrained weights
# model.load_state_dict(torch.load('LSTT_model.pth'))

# Evaluation mode
model.eval()

# Paths
input_root = "6_Sliding_Window"
output_root = "7_LSTT_Embeddings"
os.makedirs(output_root, exist_ok=True)

# For each data split
for split in ['Train', 'Test']:
    input_dir = os.path.join(input_root, split)
    output_dir = os.path.join(output_root, split)
    os.makedirs(output_dir, exist_ok=True)
    
    print(f"\nProcessing {split}...")
    
    for file in os.listdir(input_dir):
        if file.endswith(".csv"):
            stock_name = file.replace(".csv", "")
            file_path = os.path.join(input_dir, file)
            df = pd.read_csv(file_path)
            
            # Extract only Day_1 to Day_5
            try:
                X = df[['Day_1', 'Day_2', 'Day_3', 'Day_4', 'Day_5']].values.astype(np.float32)
            except KeyError:
                print(f"Skipping {file} due to missing day columns")
                continue

            if len(X.shape) != 2 or X.shape[0] == 0:
                print(f"Skipping {file} due to invalid shape {X.shape}")
                continue

            # Reshape to (batch_size=1, seq_len, input_dim)
            input_tensor = torch.tensor(X).unsqueeze(0)
            
            with torch.no_grad():
                embedding = model(input_tensor).squeeze(0).numpy()  # Shape: (64,)
            
            # Save as .npy
            np.save(os.path.join(output_dir, f"{stock_name}.npy"), embedding)

print("\n LSTT embeddings regenerated and saved as .npy files.")
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
import networkx as nx
from sklearn.preprocessing import LabelEncoder
from torch_geometric.data import Data
from torch_geometric.nn import GATConv

#  HGAT Model
class HGAT(nn.Module):
    def __init__(self, in_dim=64, hidden_dim=64, out_dim=64, heads=2):
        super(HGAT, self).__init__()
        self.gat1 = GATConv(in_dim, hidden_dim, heads=heads)
        self.gat2 = GATConv(hidden_dim * heads, out_dim, heads=1)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.elu(self.gat1(x, edge_index))
        x = self.gat2(x, edge_index)
        return x

#  Load sector mapping
sector_df = pd.read_csv("sector_mapping.csv")
stock_to_sector = dict(zip(sector_df['Stock'], sector_df['Sector']))

# Encode sectors
sector_encoder = LabelEncoder()
sector_encoded = sector_encoder.fit_transform(sector_df['Sector'])
sector_mapping = dict(zip(sector_df['Stock'], sector_encoded))

#  Graph construction
def create_hierarchical_graph(input_folder):
    G = nx.Graph()
    node_names = []
    features_dict = {}

    for file in os.listdir(input_folder):
        if file.endswith('.npy'):
            stock = file.replace('.npy', '')
            embedding = np.load(os.path.join(input_folder, file))

            if embedding is None or embedding.size == 0:
                print(f"Skipping {stock} due to empty embedding.")
                continue

            G.add_node(stock)
            G.nodes[stock]['feat'] = embedding
            features_dict[stock] = embedding
            node_names.append(stock)

    # Add sector nodes and intra-sector edges
    unique_sectors = set(sector_mapping.values())
    for sector_id in unique_sectors:
        sector_node = f"SECTOR_{sector_id}"
        G.add_node(sector_node)
        for stock, s_id in sector_mapping.items():
            if s_id == sector_id and stock in G:
                G.add_edge(stock, sector_node)
                G.add_edge(sector_node, stock)

# ðŸ”— Add inter-sector edges (newly added)
    sector_nodes = [f"SECTOR_{s_id}" for s_id in unique_sectors]
    for i in range(len(sector_nodes)):
        for j in range(i + 1, len(sector_nodes)):
            G.add_edge(sector_nodes[i], sector_nodes[j])
            G.add_edge(sector_nodes[j], sector_nodes[i])


    # Mapping nodes to indices
    name_to_idx = {name: idx for idx, name in enumerate(G.nodes())}
    edge_index = torch.tensor([[name_to_idx[u], name_to_idx[v]] for u, v in G.edges()], dtype=torch.long).t()

    # Construct feature matrix
    example_feat = next((f for f in features_dict.values() if f is not None), None)
    if example_feat is None:
        raise ValueError("âŒ No valid stock embeddings found!")

    x = []
    for node in G.nodes:
        if 'feat' in G.nodes[node]:
            x.append(G.nodes[node]['feat'])
        else:
            x.append(np.zeros_like(example_feat))

    x = torch.tensor(np.stack(x), dtype=torch.float)
    return G, node_names, x, edge_index, name_to_idx

#  HGAT Training
def train_hgat(G, x, edge_index, node_names, num_epochs=100):
    data = Data(x=x, edge_index=edge_index)
    model = HGAT(in_dim=x.size(1), hidden_dim=64, out_dim=64)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()
        out = model(data)
        loss = ((out - x) ** 2).mean()
        loss.backward()
        optimizer.step()
        if epoch % 10 == 0:
            print(f"Epoch {epoch:03d} | Loss: {loss.item():.4f}")

    return out.detach(), {name: idx for idx, name in enumerate(G.nodes())}

#  Save HGAT embeddings
def save_embeddings(output_dir, node_names, name_to_idx, embeddings):
    os.makedirs(output_dir, exist_ok=True)
    for name in node_names:
        if name.startswith("SECTOR_"):
            continue
        idx = name_to_idx[name]
        np.save(os.path.join(output_dir, f"{name}.npy"), embeddings[idx].numpy())

#  Run for each split
LSTT_FOLDER = "7_LSTT_Embeddings"
HGAT_FOLDER = "8_HGAT_Embeddings"
os.makedirs(HGAT_FOLDER, exist_ok=True)

for split in ['Train', 'Test']:
    print(f"\n Processing {split} set...")
    input_folder = os.path.join(LSTT_FOLDER, split)
    output_folder = os.path.join(HGAT_FOLDER, split)

    G, node_names, x, edge_index, name_to_idx = create_hierarchical_graph(input_folder)
    embeddings, name_to_idx = train_hgat(G, x, edge_index, node_names)
    save_embeddings(output_folder, node_names, name_to_idx, embeddings)

print("\n All HGAT embeddings saved!")
```

```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GATConv
import networkx as nx

# HGAT Model
class HGAT(nn.Module):
    def __init__(self, in_dim=64, hidden_dim=64, out_dim=64, heads=2):
        super(HGAT, self).__init__()
        self.gat1 = GATConv(in_dim, hidden_dim, heads=heads)
        self.gat2 = GATConv(hidden_dim * heads, out_dim, heads=1)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = F.elu(self.gat1(x, edge_index))
        x = self.gat2(x, edge_index)
        return x

# Graph construction without sector info
def create_graph(input_folder):
    G = nx.Graph()
    node_names = []
    features_dict = {}

    for file in os.listdir(input_folder):
        if file.endswith('.npy'):
            stock = file.replace('.npy', '')
            embedding = np.load(os.path.join(input_folder, file))

            if embedding is None or embedding.size == 0:
                print(f"Skipping {stock} due to empty embedding.")
                continue

            G.add_node(stock)
            G.nodes[stock]['feat'] = embedding
            features_dict[stock] = embedding
            node_names.append(stock)

    # Add edges based on some criteria, for now, we connect all nodes (fully connected graph)
    for i, stock_1 in enumerate(node_names):
        for stock_2 in node_names[i+1:]:
            G.add_edge(stock_1, stock_2)

    # Mapping nodes to indices
    name_to_idx = {name: idx for idx, name in enumerate(G.nodes())}
    edge_index = torch.tensor([[name_to_idx[u], name_to_idx[v]] for u, v in G.edges()], dtype=torch.long).t()

    # Construct feature matrix
    example_feat = next((f for f in features_dict.values() if f is not None), None)
    if example_feat is None:
        raise ValueError("âŒ No valid stock embeddings found!")

    x = []
    for node in G.nodes:
        if 'feat' in G.nodes[node]:
            x.append(G.nodes[node]['feat'])
        else:
            x.append(np.zeros_like(example_feat))

    x = torch.tensor(np.stack(x), dtype=torch.float)
    return G, node_names, x, edge_index, name_to_idx

# HGAT Training
def train_hgat(G, x, edge_index, node_names, num_epochs=100):
    data = Data(x=x, edge_index=edge_index)
    model = HGAT(in_dim=x.size(1), hidden_dim=64, out_dim=64)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()
        out = model(data)
        loss = ((out - x) ** 2).mean()  # Mean Squared Error loss
        loss.backward()
        optimizer.step()
        if epoch % 10 == 0:
            print(f"Epoch {epoch:03d} | Loss: {loss.item():.4f}")

    return out.detach(), {name: idx for idx, name in enumerate(G.nodes())}

# Save HGAT embeddings
def save_embeddings(output_dir, node_names, name_to_idx, embeddings):
    os.makedirs(output_dir, exist_ok=True)
    for name in node_names:
        idx = name_to_idx[name]
        np.save(os.path.join(output_dir, f"{name}.npy"), embeddings[idx].numpy())

# Main process
LSTT_FOLDER = "7_LSTT_Embeddings"
HGAT_FOLDER = "10_No_sector_HGAT_Embeddings"
os.makedirs(HGAT_FOLDER, exist_ok=True)

for split in ['Train', 'Test']:
    print(f"\nProcessing {split} set...")

    input_folder = os.path.join(LSTT_FOLDER, split)
    output_folder = os.path.join(HGAT_FOLDER, split)

    G, node_names, x, edge_index, name_to_idx = create_graph(input_folder)
    embeddings, name_to_idx = train_hgat(G, x, edge_index, node_names)
    save_embeddings(output_folder, node_names, name_to_idx, embeddings)

print("\nAll HGAT embeddings saved!")
```

```python
# Evaluate model on test set
model.eval()
with torch.no_grad():
    predicted_scores = model(X_test).numpy()

# Prepare results
results = []
for stock, score, ret, mov in zip(stock_names_test, predicted_scores, y_test_return, y_test_movement):
    results.append({
        "stock": stock,
        "score": float(score),
        "return_ratio": float(ret),
        "movement": int(mov)
    })

# Sort by predicted score
sorted_results = sorted(results, key=lambda x: x["score"], reverse=True)

# Extract values
y_true_movements = [r["movement"] for r in sorted_results]
y_pred_movements = [1 if r["score"] > 0 else 0 for r in sorted_results]
y_true_returns = [r["return_ratio"] for r in sorted_results]
y_pred_returns = [r["score"] for r in sorted_results]
stock_names = [r["stock"] for r in sorted_results]

# Metrics
mae = mean_absolute_error(y_true_returns, y_pred_returns)
acc = accuracy_score(y_true_movements, y_pred_movements)

# True MRR@K
def compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    df_sorted_pred = df.sort_values("pred", ascending=False).reset_index(drop=True)
    reciprocal_ranks = []
    for true_stock in top_k_true:
        try:
            rank = df_sorted_pred[df_sorted_pred["stock"] == true_stock].index[0]
            reciprocal_ranks.append(1.0 / (rank + 1))
        except IndexError:
            reciprocal_ranks.append(0.0)
    return sum(reciprocal_ranks) / k

# IRR@K
def compute_irr(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true_return = df.sort_values("true", ascending=False).head(k)["true"].sum()
    top_k_pred_return = df.sort_values("pred", ascending=False).head(k)["true"].sum()
    return top_k_true_return - top_k_pred_return

# Precision@K
def compute_true_precision(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_pred = df.sort_values("pred", ascending=False).head(k)["stock"].tolist()
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    intersection = set(top_k_pred).intersection(set(top_k_true))
    return len(intersection) / k

# Final Evaluation
print("\nFinal Evaluation on Test Set\n")
print(f"MAE: {mae:.4f}")
print(f"ACC: {acc:.4f}")
for k in K_list:
    mrr_k = compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k)
    precision_k = compute_true_precision(y_true_returns, y_pred_returns, stock_names, k)
    irr_k = compute_irr(y_true_returns, y_pred_returns, stock_names, k)
    print(f"[K={k}] MRR@{k}: {mrr_k:.4f} | Precision@{k}: {precision_k:.4f} | IRR@{k}: {irr_k:.4f}")
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.metrics import mean_absolute_error, accuracy_score

# Embedding paths
embedding_paths = {
    "LSTT": "7_LSTT_Embeddings",
    "HGAT": "8_HGAT_Embeddings",
    "HGAT_NoSectors": "10_No_Sector_HGAT_Embeddings"
}

# Ground truth paths
ground_truth_train = "6_Sliding_Window/Train"
ground_truth_test = "true_results.csv"
K_list = [10]

# MLP model
class StockMLP(nn.Module):
    def __init__(self, input_dim=64, hidden_dim=64):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, 1)
    def forward(self, x):
        x = self.relu(self.fc1(x))
        return self.fc2(x).squeeze()

# Load train data
def load_data_train(fused_path, ground_truth_path):
    X, y_return, y_movement, stock_names = [], [], [], []
    for file in os.listdir(fused_path):
        if not file.endswith('.npy'):
            continue
        stock = file.replace('.npy', '')
        embedding = np.load(os.path.join(fused_path, file))
        gt_file = os.path.join(ground_truth_path, f"{stock}.csv")
        if not os.path.exists(gt_file):
            continue
        df = pd.read_csv(gt_file)
        X.append(embedding)
        y_return.append(float(df['Return_Ratio'].iloc[-1]))
        y_movement.append(int(df['Movement'].iloc[-1]))
        stock_names.append(stock)
    return torch.tensor(np.array(X), dtype=torch.float32), torch.tensor(y_return), torch.tensor(y_movement), stock_names

# Load test data
def load_data_test(fused_path, ground_truth_csv):
    df_gt = pd.read_csv(ground_truth_csv)
    X, y_return, y_movement, stock_names = [], [], [], []
    for file in os.listdir(fused_path):
        if not file.endswith('.npy'):
            continue
        stock = file.replace('.npy', '')
        embedding = np.load(os.path.join(fused_path, file))
        row = df_gt[df_gt['stock'] == stock]
        if row.empty:
            continue
        return_ratio = float(row['return_ratio'].values[0])
        movement = int(row['movement'].values[0])
        X.append(embedding)
        y_return.append(return_ratio)
        y_movement.append(movement)
        stock_names.append(stock)
    return torch.tensor(np.array(X), dtype=torch.float32), torch.tensor(y_return), torch.tensor(y_movement), stock_names

# Metric functions
def compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    df_sorted_pred = df.sort_values("pred", ascending=False).reset_index(drop=True)
    reciprocal_ranks = []
    for true_stock in top_k_true:
        try:
            rank = df_sorted_pred[df_sorted_pred["stock"] == true_stock].index[0]
            reciprocal_ranks.append(1.0 / (rank + 1))
        except IndexError:
            reciprocal_ranks.append(0.0)
    return sum(reciprocal_ranks) / k

def compute_irr(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true_return = df.sort_values("true", ascending=False).head(k)["true"].sum()
    top_k_pred_return = df.sort_values("pred", ascending=False).head(k)["true"].sum()
    return top_k_true_return - top_k_pred_return

def compute_true_precision(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_pred = df.sort_values("pred", ascending=False).head(k)["stock"].tolist()
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    intersection = set(top_k_pred).intersection(set(top_k_true))
    return len(intersection) / k

# Train and Evaluate
def train_and_evaluate(train_path, test_path, name):
    X_train, y_train_return, y_train_movement, _ = load_data_train(os.path.join(train_path, "Train"), ground_truth_train)
    X_test, y_test_return, y_test_movement, stock_names_test = load_data_test(os.path.join(test_path, "Test"), ground_truth_test)

    model = StockMLP(input_dim=64)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    loss_fn = nn.MSELoss()

    for epoch in range(100):
        model.train()
        optimizer.zero_grad()
        preds = model(X_train)
        loss = loss_fn(preds, y_train_return)
        loss.backward()
        optimizer.step()

    # Eval
    model.eval()
    with torch.no_grad():
        predicted_scores = model(X_test).numpy()

    results = []
    for stock, score, ret, mov in zip(stock_names_test, predicted_scores, y_test_return, y_test_movement):
        results.append({
            "stock": stock,
            "score": float(score),
            "return_ratio": float(ret),
            "movement": int(mov)
        })

    sorted_results = sorted(results, key=lambda x: x["score"], reverse=True)
    y_true_movements = [r["movement"] for r in sorted_results]
    y_pred_movements = [1 if r["score"] > 0 else 0 for r in sorted_results]
    y_true_returns = [r["return_ratio"] for r in sorted_results]
    y_pred_returns = [r["score"] for r in sorted_results]
    stock_names = [r["stock"] for r in sorted_results]

    # Print Metrics
    print(f"\n Ablation: {name}")
    print(f"MAE: {mean_absolute_error(y_true_returns, y_pred_returns):.4f}")
    print(f"ACC: {accuracy_score(y_true_movements, y_pred_movements):.4f}")
    for k in K_list:
        mrr_k = compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k)
        precision_k = compute_true_precision(y_true_returns, y_pred_returns, stock_names, k)
        irr_k = compute_irr(y_true_returns, y_pred_returns, stock_names, k)
        print(f"[K={k}] MRR@{k}: {mrr_k:.4f} | Precision@{k}: {precision_k:.4f} | IRR@{k}: {irr_k:.4f}")

# Run Ablation
for name, path in embedding_paths.items():
    train_and_evaluate(path, path, name)
```

```python
import os
import pandas as pd
import numpy as np
import torch
from datetime import datetime, timedelta


# Paths
EMBEDDINGS_PATH = "9_Fused_Embeddings/Test"
GROUND_TRUTH_CSV = "true_results.csv"
MODEL_PATH = "trained_model.pth"
OUTPUT_DIR = "Daily_Predictions"
os.makedirs(OUTPUT_DIR, exist_ok=True)

class StockMLP(nn.Module):
    def __init__(self, input_dim=128, hidden_dim=64):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        return self.fc2(x).squeeze()
# Load trained model
model = StockMLP(input_dim=128)
model.load_state_dict(torch.load(MODEL_PATH))
model.eval()

# Load true results
df_true = pd.read_csv(GROUND_TRUTH_CSV)
stock_to_true = {
    row['stock']: (row['return_ratio'], row['movement'])
    for _, row in df_true.iterrows()
}

# Load all embeddings
all_embeddings = {}
for file in os.listdir(EMBEDDINGS_PATH):
    if file.endswith(".npy"):
        stock = file.replace(".npy", "")
        all_embeddings[stock] = np.load(os.path.join(EMBEDDINGS_PATH, file))

# Predict once for all stocks
stock_preds = []
for stock, embedding in all_embeddings.items():
    x = torch.tensor(embedding, dtype=torch.float32).unsqueeze(0)
    pred_return = model(x).item()
    if stock in stock_to_true:
        actual_return, movement = stock_to_true[stock]
        stock_preds.append({
            "Stock": stock,
            "Predicted_Return_Ratio": pred_return,
            "Actual_Return_Ratio": actual_return,
            "Movement": movement
        })

# Rank stocks by predicted return
stock_preds = sorted(stock_preds, key=lambda x: x["Predicted_Return_Ratio"], reverse=True)
for i, entry in enumerate(stock_preds):
    entry["Rank"] = i + 1

# Generate trading days
start_date = datetime.strptime("2025-01-11", "%Y-%m-%d")
end_date = datetime.strptime("2025-03-22", "%Y-%m-%d")

def is_weekday(date):
    return date.weekday() < 5  # 0=Monday, 4=Friday

current_date = start_date
while current_date <= end_date:
    if is_weekday(current_date):
        file_name = current_date.strftime("%Y-%m-%d") + ".csv"
        full_path = os.path.join(OUTPUT_DIR, file_name)
        rows = [
            {
                "Date": current_date.strftime("%Y-%m-%d"),
                **entry
            } for entry in stock_preds
        ]
        df_out = pd.DataFrame(rows)
        df_out.to_csv(full_path, index=False)
        print(f"âœ… Saved: {file_name}")
    current_date += timedelta(days=1)
 
```

```python
import pandas as pd

# Load predictions
df = pd.read_csv("predicted_results.csv")

# Evaluation Ks
K_list = [5, 10, 20]

# Initialize results
results = {k: {"MRR": 0.0, "Precision": 0.0, "IRR": 0.0} for k in K_list}
dates = df["Date"].unique()
total_days = len(dates)

# Metric functions
def compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    df_sorted_pred = df.sort_values("pred", ascending=False).reset_index(drop=True)
    reciprocal_ranks = []
    for true_stock in top_k_true:
        try:
            rank = df_sorted_pred[df_sorted_pred["stock"] == true_stock].index[0]
            reciprocal_ranks.append(1.0 / (rank + 1))
        except IndexError:
            reciprocal_ranks.append(0.0)
    return sum(reciprocal_ranks) / k

def compute_irr(y_true_returns, y_pred_returns, k):
    df = pd.DataFrame({
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_true_return = df.sort_values("true", ascending=False).head(k)["true"].sum()
    top_k_pred_return = df.sort_values("pred", ascending=False).head(k)["true"].sum()
    return abs(top_k_pred_return - top_k_true_return)

def compute_true_precision(y_true_returns, y_pred_returns, stock_names, k):
    df = pd.DataFrame({
        "stock": stock_names,
        "true": y_true_returns,
        "pred": y_pred_returns
    })
    top_k_pred = df.sort_values("pred", ascending=False).head(k)["stock"].tolist()
    top_k_true = df.sort_values("true", ascending=False).head(k)["stock"].tolist()
    intersection = set(top_k_pred).intersection(set(top_k_true))
    return len(intersection) / k

# Evaluate day by day
for date in dates:
    day_df = df[df["Date"] == date].copy()
    y_true_returns = day_df["Actual_Return_Ratio"].tolist()
    y_pred_returns = day_df["Predicted_Return_Ratio"].tolist()
    stock_names = day_df["Stock"].tolist()
    
    for k in K_list:
        mrr_k = compute_true_mrr(y_true_returns, y_pred_returns, stock_names, k)
        precision_k = compute_true_precision(y_true_returns, y_pred_returns, stock_names, k)
        irr_k = compute_irr(y_true_returns, y_pred_returns, k)

        results[k]["MRR"] += mrr_k
        results[k]["Precision"] += precision_k
        results[k]["IRR"] += irr_k

# Average over all trading days
for k in K_list:
    results[k]["MRR"] /= total_days
    results[k]["Precision"] /= total_days
    results[k]["IRR"] /= total_days

# Print final results
print(f"{'K':<5}{'MRR':>10} {'Precision':>12} {'IRR':>10}")
for k in K_list:
    print(f"k={k:<3} {results[k]['MRR']:.4f}     {results[k]['Precision']:.4f}     {abs(results[k]['IRR']):.4f}")
```

```python
import pandas as pd

# Load daily predictions
df = pd.read_csv("predicted_results.csv")

# Define evaluation function
def compute_metrics(df, k_values=[5, 10, 20]):
    results = []
    unique_dates = df["Date"].unique()
    
    for instance_index, date in enumerate(unique_dates):
        daily_df = df[df["Date"] == date].sort_values(by="Predicted_Return_Ratio", ascending=False)
        
        for k in k_values:
            top_k = daily_df.head(k)
            relevant = top_k[top_k["Movement"] == 1]

            # MRR@K
            mrr = 0
            for rank, (_, row) in enumerate(top_k.iterrows(), 1):
                if row["Movement"] == 1:
                    mrr = 1.0 / rank
                    break

            # Precision@K
            precision = len(relevant) / k

            # IRR@K: Sum of actual returns in top-k
            irr = top_k["Actual_Return_Ratio"].abs().sum()

            results.append({
                "k": k,
                "Instance_Index": instance_index,
                "MRR": mrr,
                "Precision": precision,
                "IRR": irr
            })

    return pd.DataFrame(results)

# Run evaluation
metric_df = compute_metrics(df, k_values=[5, 10, 20])

# Save
metric_df.to_csv("instance_metrics.csv", index=False)
print("âœ… Saved instance-wise metrics to 'instance_metrics.csv'")
```
